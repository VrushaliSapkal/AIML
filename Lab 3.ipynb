```python
import matplotlib.pyplot as plt
import pandas as pd
sa =  pd.read_csv('medal.csv')
sa.head()
```

```python
country_data = sa["country"]
medal_data = sa["gold_medal"]
colors = ["red", "yellow", "green", "orange", "hotpink"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Gold medal achievements of five most successful\n"+"countries in 2018 Summer Olympics")
plt.show()

```
