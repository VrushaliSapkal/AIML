```python
X=df.drop(columns='quality')
X
```

```python
Y=df['quality']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=355)
```

```python
X_train
```

```python
from sklearn import tree
classifier=tree.DecisionTreeClassifier() #model
```

```python
classifier.fit(X_train,Y_train) #fit the model
```

```python
plt.figure(figsize=(15,10)) #indicates the size of the figure
tree.plot_tree(classifier,filled=True) #plot the tree
```

```python
classifier=tree.DecisionTreeClassifier(max_depth=3) #size of tree upto 2 limits
classifier.fit(X_train,Y_train) #fit the model
plt.figure(figsize=(15,10)) #indicates the size of the figure
tree.plot_tree(classifier,filled=True) #plot the tree
```

```python
classifier.score(X_test,Y_test)
```

```python
classifier.score(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(Y_test,Y_pred)
```

```python
#GRID SEARCH CV
```

```python
grid_param={'criterion':['gini','entropy'], 'max_depth':range(2,32,1), 'min_samples_leaf':range(1,10,1),'min_samples_split':range(2,10,1), 'splitter':['best','random']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search=GridSearchCV(estimator=classifier,param_grid=grid_param,cv=5,n_jobs=-1) #model
```

```python
grid_search.fit(X_train,Y_train)
```

```python
best_para=grid_search.best_params_
print(best_para)
```
