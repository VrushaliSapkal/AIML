```python
#Bar plot
```

```python
x=np.array(['Apple','Banana','Cherry'])
y=np.array([10,20,30])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['Apple','Banana','Cherry'])
y=np.array([10,20,30])
plt.bar(x,y,color='Hotpink',width=0.5)                        #takes width not height 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Apple','Banana','Cherry'])
y=np.array([10,20,30])                                     #takes haight not width
plt.barh(x,y,color='Green',height=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(170,10,250)
print(x)
plt.hist(x)
plt.show()
```

```python
x=np.array([10,20,20,30,40,50,50,50,60])
plt.hist(x)
plt.show()
```

```python
x=np.array([10,20,30,2,3,4,30])
plt.hist(x)
plt.show()
```

```python
x=np.array([10,20,20,30,40,50,50,50,60])
y=np.array([10,20,30,2,3,4,30])
plt.hist(x)
plt.hist(y)               #Don't take plt.hist(x,y),give san error
plt.show()
```

```python
#Pie Plot
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([10,20,30,40,50])
plt.pie(y)
plt.show()
```

```python
y=np.array([10,20,30,40,50])
mylabels=['Apple','Banana','Cherry','Jamun','Chiku']
plt.pie(y,labels=mylabels)
plt.legend(title='Fruits')                    #
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
type(iris)
```

```python
print(list((iris.feature_names)))
```

```python
print(iris.target)            #prints target columns
```

```python
iris.data           #features names
```

```python
df=pd.DataFrame(iris.data)      #4 columns will be return into df
print(df)
```

```python
df.columns=iris.feature_names
print(df)
```

```python
df['label']=iris.target        #to give heading & to add new columns 
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
plt.legend(title="xyz")
```
