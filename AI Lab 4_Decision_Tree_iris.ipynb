```python
#07/05/2022
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
type(iris)
```

```python
iris
```

```python
from sklearn import tree
```

```python
classifier=tree.DecisionTreeClassifier() #model is ready
```

```python
classifier.fit(iris.data,iris.target) #fitting the model
```

```python
plt.figure(figsize=(15,10)) #indicates the size of the figure
tree.plot_tree(classifier) #plot the tree
```

```python
plt.figure(figsize=(15,10)) #indicates the size of the figure
tree.plot_tree(classifier,filled=True)
```

```python
classifier=tree.DecisionTreeClassifier(max_depth=2) #size of tree upto 2 limits
classifier.fit(iris.data,iris.target) 
plt.figure(figsize=(15,10)) #indicates the size of the figure
tree.plot_tree(classifier) #plot the tree
```

```python
classifier.score(iris.data,iris.target)
```
