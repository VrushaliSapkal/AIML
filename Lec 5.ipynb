```python
x_point=np.array([0,1,2,3])
y_point=np.array([3,8,1,10])
x1_point=np.array([0,1,2,3])
y1_point=np.array([6,2,7,11])
plt.plot(x_point,y_point,x1_point,y1_point)
plt.title('ShradAlu')
plt.xlabel("Shrad")
plt.ylabel("Alu")
plt.show()
```

```python
x_point=np.array([0,1,2,3])
y_point=np.array([3,8,1,10])
x1_point=np.array([0,1,2,3])
y1_point=np.array([6,2,7,11])
plt.plot(x_point,y_point,x1_point,y1_point)
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':20}
plt.title('ShradAlu',fontdict=font1)
plt.ylabel('Alu',fontdict=font2)
plt.xlabel('Shrad',fontdict=font2)
plt.xlabel("Shrad")
plt.ylabel("Alu")
plt.show()
```

```python
x_point=np.array([0,1,2,3])
y_point=np.array([3,8,1,10])
x1_point=np.array([0,1,2,3])
y1_point=np.array([6,2,7,11])
plt.plot(x_point,y_point,x1_point,y1_point)
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':20}
plt.title('ShradAlu',fontdict=font1)
plt.ylabel('Alu',fontdict=font2)
plt.xlabel('Shrad',fontdict=font2)
plt.xlabel("Shrad")
plt.ylabel("Alu")
plt.grid(color='green',linestyle='--',linewidth=0.5)
plt.show()
```
